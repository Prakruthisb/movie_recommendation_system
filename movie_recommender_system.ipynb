- **for example**:
'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]' **->** i need to convert this to ['Action','Adventure','Fantasy','Science Fiction']

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
def convert_cast(obj):
    L = []
    counter = 0 
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1 
        else:
            break 
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
def convert_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
import requests
import json
import time
import pandas as pd

BASE_URL = "https://caching.graphql.imdb.com/"
PAGE_SIZE = 25

HEADERS = {
    'accept': 'application/graphql+json, application/json',
    'content-type': 'application/json',
    'origin': 'https://www.imdb.com'
}

def get_variables(movie_id, after_cursor):
    variables = {
        "const": movie_id,
        "filter": {},
        "first": PAGE_SIZE,
        "sort": {"by": "HELPFULNESS_SCORE", "order": "DESC"}
    }
    if after_cursor:
        variables["after"] = after_cursor
    return variables


def fetch_page(movie_id, after_cursor):
    payload = {
        "query": """
        query TitleReviewsRefine($const: ID!, $filter: ReviewsFilter, $first: Int!, $sort: ReviewsSort, $after: ID) {
          title(id: $const) {
            reviews(filter: $filter, first: $first, sort: $sort, after: $after) {
              edges {
                node {
                  text {
                    originalText {
                      plainText
                    }
                  }
                }
              }
              pageInfo {
                hasNextPage
                endCursor
              }
            }
          }
        }""",
        "operationName": "TitleReviewsRefine",
        "variables": get_variables(movie_id, after_cursor)
    }

    response = requests.post(BASE_URL, headers=HEADERS, json=payload)
    response.raise_for_status()
    return response.json()


def get_all_review_text(movie_id, limit=1000):
    after_cursor = None
    texts = []

    # Each page gives 25 reviews → 1000 reviews ≈ 40 pages
    for _ in range((limit // PAGE_SIZE) + 5):  
        if len(texts) >= limit:
            break

        data = fetch_page(movie_id, after_cursor)
        edges = data["data"]["title"]["reviews"]["edges"]

        for review in edges:
            text = review["node"]["text"]["originalText"]["plainText"]
            texts.append(text)

            if len(texts) >= limit:
                break

        page_info = data["data"]["title"]["reviews"]["pageInfo"]
        if not page_info["hasNextPage"]:
            break

        after_cursor = page_info["endCursor"]
        time.sleep(0.4)

    # Return as ONE long string (your requirement)
    return " ".join(texts)



def get_reviews_dataframe(movie_id):
    combined_text = get_all_review_text(movie_id)
    return combined_text 
        
        
# movie_ids = ["tt2379713"]

# df = get_reviews_dataframe(movie_ids)

# print(df) 
```

```python
import requests
import time
import pandas as pd
from concurrent.futures import ThreadPoolExecutor, as_completed

BASE_URL = "https://caching.graphql.imdb.com/"
PAGE_SIZE = 25
REVIEW_LIMIT = 200  # you want 200 reviews per movie

HEADERS = {
    'accept': 'application/graphql+json, application/json',
    'content-type': 'application/json',
    'origin': 'https://www.imdb.com'
}


# ---------- RETRY SAFE REQUEST ----------
def fetch_page_with_retry(payload, max_retries=6):
    for attempt in range(max_retries):
        try:
            response = requests.post(
                BASE_URL,
                headers=HEADERS,
                json=payload,
                timeout=20
            )
            response.raise_for_status()
            return response.json()

        except Exception:
            wait = 2 ** attempt  # exponential backoff
            time.sleep(wait)

    return None  # failed even after retry


# ---------- FETCH 200 REVIEWS ----------
def get_reviews(movie_id, limit=REVIEW_LIMIT):
    after_cursor = None
    texts = []
    pages = (limit // PAGE_SIZE) + 3    # around 11 pages

    for _ in range(pages):

        # payload for IMDB GraphQL
        payload = {
            "query": """
            query TitleReviewsRefine($const: ID!, $filter: ReviewsFilter, $first: Int!, $sort: ReviewsSort, $after: ID) {
              title(id: $const) {
                reviews(filter: $filter, first: $first, sort: $sort, after: $after) {
                  edges {
                    node {
                      text {
                        originalText {
                          plainText
                        }
                      }
                    }
                  }
                  pageInfo {
                    hasNextPage
                    endCursor
                  }
                }
              }
            }""",
            "operationName": "TitleReviewsRefine",
            "variables": {
                "const": movie_id,
                "filter": {},
                "first": PAGE_SIZE,
                "sort": {"by": "HELPFULNESS_SCORE", "order": "DESC"},
                **({"after": after_cursor} if after_cursor else {})
            }
        }

        data = fetch_page_with_retry(payload)
        if data is None:
            break

        edges = data["data"]["title"]["reviews"]["edges"]

        for r in edges:
            text = r["node"]["text"]["originalText"]["plainText"]
            texts.append(text)
            if len(texts) >= limit:
                return " ".join(texts)

        page = data["data"]["title"]["reviews"]["pageInfo"]

        if not page["hasNextPage"]:
            break

        after_cursor = page["endCursor"]

        time.sleep(0.25)  # small delay → avoids IMDb blocking

    return " ".join(texts)


# ---------- PARALLEL SCRAPER ----------
def fetch_for_movie(movie_id):
    try:
        return movie_id, get_reviews(movie_id)
    except:
        return movie_id, None


def scrape_all(movies_df):
    results = {}

    with ThreadPoolExecutor(max_workers=10) as executor:
        futures = {
            executor.submit(fetch_for_movie, imdb_id): imdb_id
            for imdb_id in movies_df["imdbID"]
        }

        for future in as_completed(futures):
            mid, text = future.result()
            results[mid] = text

    movies_df["reviews"] = movies_df["imdbID"].map(results)
    return movies_df
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="allenai/led-base-16384")
sentiment_analyzer = pipeline("sentiment-analysis")

def summarize_in_chunks(text, chunk_size=3000):
    chunks = []
    for i in range(0, len(text), chunk_size):
        chunks.append(text[i:i+chunk_size])
    return chunks

text = movies['reviews'][0]

# Step 1: chunk summarization
chunk_summaries = []
for chunk in summarize_in_chunks(text):
    s = summarizer(chunk, max_length=300, min_length=150, do_sample=False)[0]["summary_text"]
    chunk_summaries.append(s)

combined_summary = " ".join(chunk_summaries)

# Step 2: RE-SUMMARIZE to make it < 500 tokens
final_summary = summarizer(
    combined_summary,
    max_length=400,
    min_length=200,
    do_sample=False
)[0]["summary_text"]

# Step 3: Sentiment analysis on final short summary
result = sentiment_analyzer(final_summary)[0]
print("\nSentiment:", result)
```
